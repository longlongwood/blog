```python
import selenium
from selenium import webdriver
from lxml import etree
browser = webdriver.Chrome()
browser.get('http://www.autohome.com')
html = browser.page_source
parser = etree.fromstring(html)
root = parser.getroottree()
citydiv = root.xpath('//*[@id="auto-header-citypop-citylist"]')
for elea in citydiv[0].iter('{*}dl'):
    for i in range(len(elea.xpath('descendant::*[@class="tx"]/text()'))):
        #print(elea.xpath('descendant::*[@class="tx"]/text()')[i])
        for cityele in elea.xpath('child::*[local-name()="dd"]')[i]:
            print(elea.xpath('descendant::*[@class="tx"]/text()')[i]+cityele.text+','+cityele.attrib['data-key'])
```
